# Extraer noticias de el tiempo

El siguiente código muestra una rutina que nos permite obtener noticias del periódico [El Tiempo](https://www.eltiempo.com) y organizarlas en una sencilla Base de datos.

```python
import re
import operator
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen

```

```python
links = []
for i in range(2,10):
    pagina_bus=urlopen("https://www.eltiempo.com/buscar/"+str(i)+"?q=&publishedAt%5Bfrom%5D=05-04-16&publishedAt%5Buntil%5D=20-04-02&contentTypes%5B0%5D=article")## Está es una página de búsqueda genérica, si ustedes prefieren pueden cambiarla
    CodHTML_bus=pagina_bus.read().decode('utf-8')
    CodHTMLclean_bus=BeautifulSoup(CodHTML_bus)
    tabla = CodHTMLclean_bus.find_all( attrs={'class': 'title page-link'})
    for link in tabla:
        links.append('https://www.eltiempo.com'+link.get('href'))
    
```

```python
links 
```

```python
import pandas as pd
linksdata=pd.DataFrame(data=links,columns=["Enlaces"])
linksdata.to_excel("Linksguardados.xlsx")
```

```python
Noticias=[]
for noticia in links:
    pagina=urlopen(noticia)
    CodHTML=pagina.read().decode("utf-8")
    pagina_noticia=BeautifulSoup(CodHTML)
    if pagina_noticia.find( 'h1')!=None:
        titulo=pagina_noticia.find( 'h1').get_text(' ', strip=True)
    else:
        titulo=""
    if pagina_noticia.find( attrs={'class': 'info'})!=None:
        info=pagina_noticia.find( attrs={'class': 'info'}).get_text(' ', strip=True)
    else:
        info="" 
    tabla = pagina_noticia.find_all( attrs={'class': 'contenido'})
    tabla= [i.get_text(' ', strip=True) for i in tabla]
    tabla=' '.join(tabla)
    Noticias.append([titulo,info,tabla])
    
```

```python
DF=pd.DataFrame(data=Noticias,columns=["Título","info","contenido"])
DF.to_excel("Noticias.xlsx")
```
